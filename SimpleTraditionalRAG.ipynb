### Module Imports
This cell imports all the necessary Python modules and classes from various libraries that will be used throughout the notebook to construct the RAG (Retrieval Augmented Generation) pipeline. These include components for interacting with Google's generative AI, defining prompts, parsing outputs, text splitting, document loading, and vector storage.

```python
import os
import time
from getpass import getpass
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import Chroma
```

### Data Preparation: Creating a Sample Text File
This cell initializes the data that will be used as the knowledge base for the RAG system. It defines a multi-line string containing information about a fictional planet, Aethelgard. This `sample_text` is then written to a local file named `sample.txt`, which will subsequently be loaded and processed.

```python
# 2. Data Preparation: Create a sample text file about Aethelgard
sample_text = """The planet Aethelgard is a terrestrial world located in the triple-star system of Alpha Centauri Prime. It is characterized by its vibrant purple flora and massive floating islands suspended by geomagnetic anomalies. The atmosphere is rich in neon and argon, creating a perpetual twilight glow across the bioluminescent landscapes.

Technologically, the inhabitants of Aethelgard have mastered crystalline energy resonance. By harvesting vibrations from the planet's core, they power vast metropolitan hubs without any carbon footprint. Their architecture seamlessly integrates with the natural rock formations, utilizing translucent minerals that change color based on the local temperature and humidity levels.

Despite its beauty, Aethelgard faces challenges from periodic 'Ionic Storms' that sweep across the equator. these storms can disrupt electronic communications for weeks, forcing the population to rely on ancient optical telegraphy systems. Scientists are currently researching synthetic shielding made from crushed meteorites to stabilize the power grids during these high-energy atmospheric events."""

with open('sample.txt', 'w') as f:
    f.write(sample_text)
```

### Loading and Chunking Documents
This cell is crucial for processing the raw text data. It uses `TextLoader` to read the content from `sample.txt`. After loading, a `RecursiveCharacterTextSplitter` is initialized to break down the document into smaller, overlapping chunks. This chunking strategy helps ensure that relevant context can be retrieved efficiently without overwhelming the language model.

```python
# 3. Loading and Chunking: Load document and split into manageable pieces
loader = TextLoader('sample.txt')
documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_documents(documents)
```

### Generating Embeddings and Creating a Vector Store
This cell converts the text chunks into numerical representations called embeddings using `GoogleGenerativeAIEmbeddings` with the `gemini-embedding-001` model. These embeddings capture the semantic meaning of the text. Subsequently, these embedded chunks are stored in a `Chroma` vector database, enabling fast and efficient similarity search (retrieval) later in the RAG pipeline.

```python
# 4. Embeddings and Vector Store: Convert text to vectors and index them
embeddings = GoogleGenerativeAIEmbeddings(model='models/gemini-embedding-001')
vectorstore = Chroma.from_documents(documents=chunks, embedding=embeddings)
```

### LLM Initialization and Configuration
This cell initializes the large language model (LLM) that will be used for generating answers. It uses `ChatGoogleGenerativeAI` with the `gemini-2.5-flash` model. Important parameters like `max_retries` are set to handle transient network issues or rate limits, and `temperature` is set to `0` for more deterministic and factual responses.

```python
# 5. LLM Initialization: Setup the Gemini model with retry parameters
# Using max_retries to handle transient 503/429 errors
llm = ChatGoogleGenerativeAI(
    model='models/gemini-2.5-flash',
    max_retries=5,
    temperature=0
)
```

### Prompt Engineering: Defining the RAG System Prompt
This cell focuses on prompt engineering, which is crucial for guiding the LLM's behavior. It defines a `ChatPromptTemplate` that instructs the model on how to answer questions. The prompt explicitly tells the model to use the provided context, state if it doesn't know the answer, and keep responses concise (maximum three sentences).

```python
# 6. Prompt Engineering: Define the RAG system prompt
template = """You are an assistant for question-answering tasks. Use the following pieces of retrieved context to answer the question. If you don't know the answer, just say that you don't know. Use three sentences maximum and keep the answer concise.

Question: {question}

Context: {context}

Answer:"""
prompt = ChatPromptTemplate.from_template(template)
```

### LCEL Chain Construction: Building the End-to-End RAG Pipeline
This cell constructs the complete RAG (Retrieval Augmented Generation) pipeline using LangChain Expression Language (LCEL). It defines a sequence of operations: first, it retrieves relevant context from the vector store; then, it passes this context along with the user's question to the defined prompt; the output of the prompt is fed into the LLM; and finally, `StrOutputParser` extracts the string response from the LLM's output.

```python
# 7. LCEL Chain Construction: Define the end-to-end RAG pipeline
rag_chain = (
    {"context": vectorstore.as_retriever(), "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

### Execution: Running a Query and Handling Errors
This cell demonstrates how to execute a query using the constructed RAG pipeline. It defines a sample question about 'Ionic storms' and then invokes the `rag_chain` to get a response. Basic error handling is included to catch potential exceptions, such as server overload issues, and provides a user-friendly message to retry the operation.

```python
# 8. Execution: Run the query with basic error handling for server spikes
query = 'What do you know about Ionic storms?'

try:
    response = rag_chain.invoke(query)
    print(f'Query: {query}')
    print(f'\nResponse: {response}')
except Exception as e:
    print(f'An error occurred: {e}')
    print('The server might be experiencing high demand. Please wait a moment and run this cell again.')
```
